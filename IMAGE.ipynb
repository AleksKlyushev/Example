```python
import mahotas as mh
raw_dir = 'C:\\Users\\123\\Documents\\Python\\BuildingMachineLearningSystemsWithPython\\ch10\\'
image = mh.imread(raw_dir+'scene00.jpg')
```

```python
from matplotlib import pyplot as plt
import numpy as np
#plot graf in the same window :
%matplotlib inline 

plt.imshow(image)
plt.show()
```

```python
image2 = mh.colors.rgb2grey(image,dtype = np.uint8)
plt.imshow(image2)
plt.gray()
```

```python
thresh = mh.thresholding.otsu(image2)
print('порог Оцу равен : {}'.format(thresh))
plt.imshow(image2 > thresh)
```

```python
r,g,b = image.transpose(2,0,1)
r6 = mh.gaussian_filter(r,6.)
g6 = mh.gaussian_filter(g,6.)
b6 = mh.gaussian_filter(b,6.)
img6 = mh.as_rgb(r6,g6,b6)
plt.imshow(img6)
plt.show()
```

```python
im16 = mh.gaussian_filter(image2,32)
plt.imshow(im16)
plt.show()
```

```python
im10 = mh.gaussian_filter(image2,10)
thresh = mh.thresholding.otsu(image2)
print('порог Оцу равен : {}'.format(thresh))
plt.imshow(im10 > thresh)
```

```python
im = mh.demos.load('lena')
plt.imshow(im)
plt.show()
```

```python
r,g,b = im.transpose(2,0,1)
r12 = mh.gaussian_filter(r,12.)
g12 = mh.gaussian_filter(g,12.)
b12 = mh.gaussian_filter(b,12.)
im12 = mh.as_rgb(r12,g12,b12)

plt.imshow(im12)
plt.show()
```

```python
h, w = r.shape
y, x = np.mgrid[:h,:w]

y = y - h/2.
y = y / y.max()

x = x - w/2.
x = x / x.max()

c = np.exp(-2.*(x**2+y**2))
c = c - c.min()
c = c / c.ptp()
c = c[:,:,None]

ringen = mh.stretch(im*c + (1-c)*im12)
plt.imshow(ringen)
plt.show()
```

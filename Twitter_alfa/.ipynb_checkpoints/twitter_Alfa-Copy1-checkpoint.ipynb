```python
df_word_a = pd.DataFrame(vectorizer.get_feature_names(),columns=['word'])
df_word_a['count'] = q
print(len(df_word_a))
print(df_word_a.sort('count',ascending=False))
```

```python
import nltk
rus_stem = nltk.stem.SnowballStemmer('russian')
# print(s.stem('банков'))
# print(s.stem('банком'))
#переопределение функции CountVectorizer
class StemmedCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer = super(StemmedCountVectorizer,self).build_analyzer()
        return(lambda doc: (rus_stem.stem(w) for w in analyzer(doc)))
```

```python
vectorizer = StemmedCountVectorizer(min_df=1, stop_words=rus,)
v = vectorizer.fit_transform(df_A.text_new)
e = 0
zz=v.toarray()
q = zz[0]
for i in zz[1:]:
    q += i
df_word_a = pd.DataFrame(vectorizer.get_feature_names(),columns=['word'])
df_word_a['count'] = q
print(len(df_word_a))
print(df_word_a.sort('count',ascending=False))
```
